Mendefinisakan sebuah array

```python
nilai_mahasiswa= []
```

Menginputkan 10 nilai mahasiswa

```python
nilai_mahasiswa = []
for i in range(10):
    n = int(input(f"nilai mahasiswa ke-{i+1} :"))
    nilai_mahasiswa.append(n)
```

Tampilkan nilai tertinggi & terendah

```python
nilai_tertinggi = max(nilai_mahasiswa)
nilai_terendah = min(nilai_mahasiswa)

print(f"Nilai tertinggi: {nilai_tertinggi}")
print(f"Nilai terendah: {nilai_terendah}")
```

Hitung rata-rata

```python
rata_rata = sum(nilai_mahasiswa) / len(nilai_mahasiswa)
print(f"Rata-rata nilai: {rata_rata}")
```

Menghitung jumlah mahasiswa lulus dengan ketentuan nilai di atas atau sama dengan 60 lulus

```python
jumlah_lulus = 0
for nilai in nilai_mahasiswa:
    if nilai >= 60:
        jumlah_lulus += 1

print(f"Jumlah mahasiswa yang lulus: {jumlah_lulus}")
```

Membuat grafik untuk menggambarkan nilai tertinggi dan terendah

```python
import matplotlib.pyplot as plt

# Data untuk grafik
nama_nilai = ['Nilai Tertinggi', 'Nilai Terendah']
values = [nilai_tertinggi, nilai_terendah]

# Membuat bar chart
plt.figure(figsize=(7, 5))
plt.bar(nama_nilai, values, color=['skyblue', 'lightcoral'])

# Menambahkan judul dan label
plt.title('Perbandingan Nilai Tertinggi dan Terendah Mahasiswa')
plt.ylabel('Nilai')
plt.ylim(0, 100) # Asumsi nilai antara 0 dan 100

# Menampilkan nilai di atas bar
for i, val in enumerate(values):
    plt.text(i, val + 2, str(val), ha='center', va='bottom')

# Menampilkan grafik
plt.show()
```

Membuat grafik untuk menggambarkan data kelulusan

```python
import matplotlib.pyplot as plt

total_mahasiswa = len(nilai_mahasiswa)
jumlah_tidak_lulus = total_mahasiswa - jumlah_lulus

# Data untuk bar chart
labels = ['Lulus', 'Tidak Lulus']
sizes = [jumlah_lulus, jumlah_tidak_lulus]
colors = ['green', 'red']

# Membuat bar chart
plt.figure(figsize=(7, 5))
plt.bar(labels, sizes, color=colors)

# Menambahkan judul dan label
plt.title('Jumlah Mahasiswa Lulus dan Tidak Lulus')
plt.ylabel('Jumlah Mahasiswa')
plt.xlabel('Status Kelulusan')
plt.ylim(0, total_mahasiswa + 2) # Menyesuaikan batas y-axis agar angka terlihat

# Menampilkan jumlah di atas bar
for i, val in enumerate(sizes):
    plt.text(i, val + 0.5, str(val), ha='center', va='bottom')

# Menampilkan grafik
plt.show()
```
